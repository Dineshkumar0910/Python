```python
#User defined Exceptions
def SI(amount, year, rate):
    try:
        if rate>100:
            raise ValueError(rate)
        interest = (amount*year*rate)/100
        print(interest)
        return interest
    except ValueError:
        print("This is incorrect rate")

SI(1000,1,2)
SI(3442,3,102)
```
